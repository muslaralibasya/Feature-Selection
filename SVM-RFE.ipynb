```python
#Import Packages
from sklearn import svm
from sklearn.svm import SVC
import numpy as np
import pandas as pd
from sklearn.feature_selection import RFE
import timeit
```

```python
df = pd.read_csv('C:/Data Basya/Skripsi/Skripsi Basya/SKL/ISmart/Hasil Penelitian/data paper.csv')
df
```

```python
y = df.Label
x = df.drop(['ID_REF','Label'], axis=1)
x.head()
```

```python
y.head()
```

```python
# 1 = Lung Tumor
# -1 = Normal Lung

y.value_counts()
```

```python
#RFE Baru

#Fix
start = timeit.default_timer()

clf = SVC(gamma='auto', kernel='linear', random_state=55) #Train svm biasa
selector = RFE(clf, n_features_to_select=1000, step=25)
datal1 = selector.fit_transform(x,y)

stop = timeit.default_timer()
print("Time: "+"%s seconds" % (stop - start))

print(datal1.shape)
```

```python
datal1 = pd.DataFrame(datal1)
datal1
```
